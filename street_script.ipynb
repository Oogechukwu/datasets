```python
import cv2
from ultralytics import YOLO
import requests
import os

# Download video if not exists
video_path = "strreet.mp4"
if not os.path.exists(video_path):
    url = "https://github.com/Oogechukwu/datasets/raw/main/strreet.mp4"
    r = requests.get(url, stream=True)
    with open(video_path, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)

# Load YOLOv8 model
model = YOLO("yolov8n.pt")  # or your trained model

# Run video inference
results = model.predict(source=video_path, stream=True, show=False)

for r in results:
    frame = r.plot()
    cv2.imshow("YOLOv8 Stream", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
